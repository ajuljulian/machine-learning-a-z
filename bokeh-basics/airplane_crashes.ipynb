```python
import pandas as pd

df = pd.read_csv('data/Airplane_Crashes_and_Fatalities_Since_1908.csv')
```

```python
df.head()
```

```python
df['Year'] = df.apply(lambda row: int(row['Date'].split('/')[-1]), axis=1)
```

```python
df.head()
```

```python
df.groupby(['Operator', 'Year'])['Year'].describe()
```

```python
df_major_crashes = df.loc[df['Fatalities'] > 50].loc[df['Year'] > 1970]
```

```python
df_major_crashes.describe()
```

```python
df_operator_counts = pd.DataFrame(df_major_crashes.groupby(['Operator'])['Operator'].count())

#df_operator_counts = df_operator_counts.reset_index().rename(columns={})
#df_operator_counts.columns

df_operator_counts.columns
```
