Use contour plots to display 3D data on a 2D surface.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black');
```

```python
plt.contourf(X, Y, Z, 20, cmap=plt.cm.RdGy);
plt.colorbar()
```

```python
def sin_cos(x, y):
    return np.sin(x) + np.cos(y)

x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 1, 2, 3, 4])
(X, Y) = np.meshgrid(x, y)
Z = sin_cos(X, Y)
plt.contourf(X, Y, Z, 200, cmap=plt.cm.Pastel1);
```
