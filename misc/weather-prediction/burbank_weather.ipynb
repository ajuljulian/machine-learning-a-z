### Predict the Future

https://facebook.github.io/prophet/docs/quick_start.html

```python
from fbprophet import Prophet
```

```python
df_temp_daily.head()
```

```python
df_daily_temps = df_temp_daily.copy()
df_daily_temps = df_daily_temps.reset_index()
df_daily_temps = df_daily_temps.rename(columns={'DATE':'ds', 'HOURLYDRYBULBTEMPF':'y'}).copy()
```

```python
df_daily_temps.info()
```

```python
df_daily_temps.head()
```

```python
m = Prophet()
m.fit(df_daily_temps)
```

```python
future = m.make_future_dataframe(periods=18250)
```

```python
future.tail()
```

```python
forecast = m.predict(future)
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m.plot(forecast)
```
